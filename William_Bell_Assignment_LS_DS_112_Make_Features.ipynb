```python
### Write a function that can do what we have written above to any 
### string that is passsed to it.
term_rate = df['term'][2]


def str_to_int(cell_contents):
  return int(cell_contents.strip().strip(' months'))

str_to_int(term_rate)
```

```python
### is the data type correct?

type(str_to_int(term_rate))
```

```python
# .apply()to every cell in term column from str to int

df['term_integer'] = df['term_rate'].apply(str_to_int)

df.head()
```

```python
# Make a column named 'loan_status_is_great'. It should contain the integer 1 
# if 'loan_status' is "Current" or "Fully Paid." Else it should contain the integer 0.

df['loan_status_is_great'] = df[df['loan_status'] == "Current" or df['loan_status'] == "Fully Paid"]


def str_to_int(loan_status):
  return 1 if loan_status is True else 0

df['loan_status_is_great'] = df['loan_status_is_great'].apply(str_to_int)
```

```python
# Make last_pymnt_d_month and last_pymnt_d_year columns.

import pandas as pd

[col for col in df if col.endswith('_d')]

df['last_pymnt_d'] = pd.to_datetime(df['last_pymnt_d'], infer_datetime_format=True)

df['last_pymnt_d'].head().values
```

```python
df['last_pymnt_d_month'] = df['last_pymnt_d'].dt.month

df['last_pymnt_d_month']
```

```python
df['last_pymnt_d_year'] = df['last_pymnt_d'].dt.year

df['last_pymnt_d_year']
```
